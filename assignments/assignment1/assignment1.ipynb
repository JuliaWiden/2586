### Task 1

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a) 
filtered_transactions = [item for item in transactions if item['price'] > 0 and item['quantity'] >= 0]

# b)
for item in filtered_transactions:
    item['total'] = item['price'] * item['quantity']
    
# c) Grouping and computing revenue by region, storing in a dictionary
revenue_per_region = {}
for item in filtered_transactions:
    if item['region'] in revenue_per_region:
        revenue_per_region[item['region']] = revenue_per_region[item['region']] + item['total']
    else:
        revenue_per_region[item['region']] = item['total']
        
# d) Printing the region with the highest total revenue and its value using an f-string
highest_region = max(revenue_per_region, key=revenue_per_region.get)
highest_value = revenue_per_region[highest_region]

answer = f"{highest_region} has the highest total revenue at {highest_value}!"
print(answer)
```

### Task 2

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales): # defining the function
    total = sum(sales) # calculating the total sum of sales during the past 2 weeks
    average = round(sum(sales) / len(sales), 2) # finding the average and rounding to two decimals
    best_day = sales.index(max(sales)) + 1 # finding the day with the highest sales starting from the first day
    worst_day = sales.index(min(sales)) + 1 # finding the day with the lowest sales starting from the first day
    above_average_count = len([item for item in sales if item > average]) # finding the days where the sales exceeded the average
    
    return { 
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    } 
print(summarise_sales(sales))
```

### Task 3

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

```python
# a) Creating a list of students who have a passing grade (50 or above)
passing = [item['name'] for item in students if item['grade'] >= 50]
print(passing)
```

```python
# b) Looping through students and print their status based on their grade
for item in students:
    if item['grade'] >= 80:
        result = "Excellent"
    elif item['grade'] >= 50:
        result = "Pass"
    else:
        result = "Fail"
   
    print(f" {item['name']}: {result}")
```

```python
# c) Counting how many students have an attendance rate below 0.80
count = len([item for item in students if item['attendance'] < 0.80])
print(count)
```

```python
# d) Finding the student with the highest grade using a for loop
best_student = students[0] 

for item in students:
    if item['grade'] > best_student['grade']:
        best_student = item 
print(f"{best_student['name']} {best_student['grade']}")
```

```python
# e) Finding the student with the highest grade using max()
def what_grade(s):
    return s['grade']

best_student_max = max(students, key=what_grade)

print(f"{best_student_max['name']} {best_student_max['grade']}")
```

### Task 4

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}
        
    # a)
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}
    
    # b)
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            total = total + self.holdings[ticker]['shares'] * self.holdings[ticker]['price_per_share']
        return total

    # c) 
    def most_valuable(self):
        return max(self.holdings, key=lambda ticker: self.holdings[ticker]['shares'] * self.holdings[ticker]['price_per_share'])
    
    # d) 
    def __str__(self):
        return f"Owner: {self.owner}, Assets: {len(self.holdings)}, Total Value: {self.total_value()}"
    
# Putting it all together

my_portfolio = Portfolio("Julia") # Creating an instance

my_portfolio.add_asset("NVDA", 10, 167.5) # Adding 3 assets
my_portfolio.add_asset("AAPL", 5, 248.8)
my_portfolio.add_asset("MSFT", 3, 356.8)

print(my_portfolio.total_value()) # Demonstrating the methods
print(my_portfolio.most_valuable())
print(my_portfolio)
```

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

```python
# a) Creating DataFrame
traffic_data = pd.DataFrame({ 
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

traffic_data = traffic_data.set_index("date") # Setting date as index
```

```python
# b) 
traffic_data["conversions"] = (traffic_data["visitors"] * traffic_data["conversion_rate"]).round(0) # New column 1
traffic_data["daily_revenue"] = traffic_data["conversions"] * traffic_data["avg_order_value"] # New column 2

# Printing to check for accuracy
traffic_data
```
